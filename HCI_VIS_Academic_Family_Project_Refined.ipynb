```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HCI-VIS Data/For Colab/AuthorPublication-uptoVIS2020.csv')
df
```

```python
firstAuthors = df.loc[(df['AuthorOrderRank'] == "First") & (df['CareerAge'] <= 5), ["DOI", "Author"]]
firstAuthors = firstAuthors.rename(columns={"Author": "Supervisee"})
firstAuthors
```

```python
lastAuthors = df.loc[df["AuthorOrderRank"] == "Last", ["DOI", "Author"]]
lastAuthors = lastAuthors.rename(columns={"Author": "Supervisor"})
lastAuthors
```

```python
firstLastdf = firstAuthors.merge(lastAuthors, how="inner", on="DOI")
firstLastdf = firstLastdf.sort_values('Supervisor')
firstLastdf
```

```python
firstLastdfGrouped = firstLastdf.groupby(['Supervisee', 'Supervisor']).agg({
    'DOI': 'count'
}).reset_index()
firstLastdfGrouped
```

```python
firstLastdfGrouped.to_csv('VIS_supervision.csv', index=False)
```

```python
vispubMain = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HCI-VIS Data/For Colab/IEEE VIS papers 1990-2021 - Main dataset.csv')
vispubMain = vispubMain.filter(['DOI','Title'], axis=1)
vispubMain
```

```python
titlePaperID = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HCI-VIS Data/For Colab/Paper.csv')
titlePaperID = titlePaperID.filter(['Paper_ID','Title'], axis=1)
titlePaperID
```

```python
mergedDOIPaperID = vispubMain.merge(titlePaperID, how="inner", on="Title")
mergedDOIPaperID = mergedDOIPaperID.filter(['DOI', 'Paper_ID'], axis=1)
mergedDOIPaperID
```

```python
vis2020 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HCI-VIS Data/For Colab/AuthorPublication-uptoVIS2020.csv')
vis2020merged = vis2020.merge(mergedDOIPaperID, how='inner', on='DOI')
vis2020merged = vis2020merged.drop('Conference', axis=1)
vis2020merged
```

```python
biblioVIS = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HCI-VIS Data/For Colab/BiblioVIS - final.csv')
biblioVIS = biblioVIS.rename(columns={"Paper ID": "Paper_ID"})
biblioVIS
```

```python
from nameparser import HumanName
biblioVIS['Author'] = biblioVIS['Author'].astype(str)
biblioVIS['Author Last Name'] = biblioVIS['Author'].apply(lambda x: HumanName(x).last)
biblioVIS
```

```python
import re
vis2020merged['Author'] = vis2020merged['Author'].astype(str)
vis2020merged["Author Last Name"] = vis2020merged['Author'].apply(lambda s: re.sub("^\d+\s|\s\d+\s|\s\d+$", " ", s))
vis2020merged["Author Last Name"] = vis2020merged['Author Last Name'].apply(lambda x: HumanName(x).last)
vis2020merged["Author Last Name"] = vis2020merged['Author Last Name'].str.strip()

vis2020merged
```

```python
finalMergeRetry = biblioVIS.merge(vis2020merged, how="left", on=["Paper_ID", "Author Last Name"])
finalMergeRetry.sort_values("Paper_ID")
```

```python
supervision = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HCI-VIS Data/VIS_supervision.csv')
supervision
```

```python
freqSupervisor = supervision['Supervisor'].value_counts()
freqSupervisor.to_csv('Supervisor Frequency.csv', header=True)
```

```python
freqSupervisee = supervision['Supervisee'].value_counts()
freqSupervisee.to_csv('Supervisee Frequency.csv', header=True)
```

```python
import networkx as nx
G = nx.DiGraph()
```

```python
for i, row in supervision.iterrows():
    supervisor = row['Supervisor']
    supervisee = row['Supervisee']
    w = row['DOI']
    G.add_edge(supervisor, supervisee, weight = w)
```

```python
nx.write_gexf(G, "supervision.gexf")
```
